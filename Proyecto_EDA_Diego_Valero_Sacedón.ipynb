### 3. Análisis **Univariado**:

```
# 3. Análisis Univariado del conjunto de datos de riesgo y aprobación de préstamos.

# Eliminación de valores atípicos (outliers) dentro de las variables numéricas.
key_cols=['Age', 'Experience', 'LoanDuration', 'NumberOfDependents', 'CreditCardUtilizationRate', 'NumberOfOpenCreditLines', 'NumberOfCreditInquiries', 'DebtToIncomeRatio', 'PaymentHistory', 'LengthOfCreditHistory', 'SavingsAccountBalance', 'CheckingAccountBalance', 'TotalAssets', 'TotalLiabilities', 'UtilityBillsPaymentHistory', 'JobTenure', 'BaseInterestRate', 'InterestRate']

# Eliminación de valores atípicos (outliers) utilizando el Rango Intercuartílico.
def remove_outliers_iqr(data, column):
    # Conversión de la columna a datos numéricos para evitar el error de tipología de datos.
    data[column]= pd.to_numeric(data[column], errors= 'coerce')

    Q1 = data[column].quantile(0.25)  # Primer cuartil (25%)
    Q3 = data[column].quantile(0.75)  # Tercer cuartil (75%)
    IQR = Q3 - Q1                     # Rango Intercuartílico
    lower_bound = Q1 - 1.5 * IQR      # Límite inferior
    upper_bound = Q3 + 1.5 * IQR      # Límite superior

# Filtrado de datos dentro de los límites establecidos.
    filtered_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return filtered_data

# Aplicación de la eliminación de valores atípicos (outliers) a cada una de las columnas clave.
for column in key_cols:
    data = remove_outliers_iqr(data, column)
    print(f"Outliers eliminados en la variable '{column}'. Tamaño del dataset: {data.shape}")

# Verificación del nuevo tamaño en el dataset después de la eliminación de valores atípicos (outliers).
print("Dataset después de la eliminación de outliers:")
print(data.shape)
```

```
# Generación de tablas de contingencia para cada una de las variables categóricas en relación con "Estado de Aprobación del Préstamo".
categorical_cols = ['EmploymentStatus', 'EducationLevel', 'MaritalStatus', 'HomeOwnershipStatus', 'LoanPurpose']

for col in categorical_cols:
    print(f"Tabla de contingencia para {col} y LoanApproved:")

    # Generación de tabla de cálculo de frecuencias absolutas.
    contingency_table = pd.crosstab(data[col], data['LoanApproved'])
    print(contingency_table)
    print()

    # Generación de tabla de cálculo frecuencias relativas (probabilidades por fila)
    probability_table = pd.crosstab(data[col], data['LoanApproved'], normalize= 'index')
    print(f"Tabla de frecuencias relativas para {col} y LoanApproved:")
    print(probability_table)
    print()

```
